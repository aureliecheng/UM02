```python
import pandas as pd
# Matplotlib for additional customization: long conde because u need to spcidy everything
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
```

```python
df = pd.read_csv('UTSEUS-anjuke-real-estate.csv')
df.head()
```

```python
# Set themes
sns.set()
```

# Scatter plot

```python
# Recommended way
plt.figure(figsize=(12,8))
sns.lmplot(x='price', y='surface', data=df, 
          fit_reg=False, # No regression line
          hue='district')
plt.xscale('log')
plt.yscale('log')
```

# Boxplot

```python
plt.figure(figsize=(12,8))
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=df)
```

Pre-format DF

```python
districtPrice_df = df.drop(['longitude', 'latitude', 'price', 'surface', 'onesquaremeter', 'done'], axis=1)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=districtPrice_df)
```

# Violin Plot

Great to visualize distributions 

```python
plt.figure(figsize=(20,12))
# Violin plot
sns.violinplot(x='district', y='onesquaremeter', data=df)
plt.yscale('log')
```

# KDE plot

```python
# Density Plot
sns.kdeplot(df.longitude, df.latitude)
```
